이제 적절한 릿지와 라쏘에서 alpha 값을 찾아보자

```
import matplotlib.pyplot as plt
train_score = []
test_score = []
```

```
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
```

```
for alpha in alpha_list:
  ridge = Ridge(alpha=alpha)
  ridge.fit(train_scaled, train_target)
  train_score.append(ridge.score(train_scaled,train_target))
  test_score.append(ridge.score(test_scaled,test_target))
```

```
train_score
```

```
test_score
```

alpha(규제강도)를 높일수록 회귀선이 훈련데이터셋엔 덜 정확해지고 테스트데이터셋엔 더 정확해진다.

이제 그래프를 그려보자. 하지만 x값이 0.001, 0.1 1, ... 등 왼쪽이 너무 촘촘하고 오른쪽은 너무 느슨하기 때문에 log10 으로 바꿔준다.

```
plt.plot(np.log10(alpha_list),train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('알파')
plt.ylabel('정확도')
```

위에 것이 훈련 데이터셋 정확도, 아래 것이 테스트 데이터셋 정확도

최적값은 테스트정확도가 가장 높고, 훈련정확도는 보통인 -1이다.

즉 알파 = 0.1

```
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)
ridge.score(train_scaled,train_target)
```

```
ridge.score(test_scaled, test_target)
```

lr2 의 계수와 절편이 복잡함. 이 의미는 무엇인가?

LogisticRegression 은 각각의 class 에 해당하는 z값을 모두 구해냄.

첫번째, Bream 을 알아내는 함수는

[-1.17210702 -1.22527457  2.58776426  7.6850936  -0.71840643]
와
절편값 0.24263567
으로 이루어진 선형함수임.

두번째 Parki 는 마찬가지로 그 다음 행과, 두 번째 b값에 해당할 것이고.

그리고 LogisticRegression 은 마지막에 z값을 구하고 Sigmoid 함수가 아닌 Softmax 함수를 사용하여 7개의 z값을 확률로 변환한다.

시그모이드는 하나의 선형 방정식의 결과값을 0~1 사이로 변환.

Softmax 는 여러개의 선형 방정식의 출력을 0~1 사이로 압축하고,

전체 합이 1이 되도록 맞춤.
